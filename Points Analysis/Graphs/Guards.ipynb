```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression               # Imported libraries
data = pd.read_csv("Regression Model Data +Tech - 2024 Draft Combine Stats (5).csv")
data.replace("-", pd.NA, inplace=True)
data["Shuttle Run"] = pd.to_numeric(data["Shuttle Run"])
data["Three Quarter Sprint"] = pd.to_numeric(data["Three Quarter Sprint"])      # Changed datatypes to numeric to proceed with analysis
data.head()
```

```python
position_map = {
    "PG": "Guard", 
    "SG": "Guard",
    "PG-SG": "Guard",
    "SF": "Forward", 
    "PF": "Forward", 
    "C": "Center"
    }

data["POS"] = data["POS"].replace(position_map)         # Changed the format of the position to column to group players as Guards, Forwards, Centers
data.head()
```

```python
grouped = data["POS"].value_counts()
grouped
```

### **All graphs below will compare the variables with Points Per Game for only Guards**

```python
guards = data[data["POS"] == "Guard"]
plt.figure(figsize= (7,4))
x = guards[["PPG"]]
y = guards["Lane Agility Time"]
model = LinearRegression()
model.fit(x,y)
predictions = model.predict(x)
plt.scatter(x,y, color = "green")
plt.plot(x, predictions, color = "black")
plt.xlabel("Points Per Game")
plt.ylabel("Lane Agility Time")
plt.title("Points v Lane Agility Time")
plt.ylim(10.25,13)
plt.show()
```
