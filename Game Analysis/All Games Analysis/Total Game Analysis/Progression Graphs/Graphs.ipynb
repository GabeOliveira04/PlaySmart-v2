```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

Advanced = pd.read_csv("advanced_data.csv")
Basic = pd.read_csv("basic_data.csv")
```

```python
counter = range(1, len(Advanced) + 1)
X = [[x] for x in counter] 
y = Advanced["PPFGA"].values

model = LinearRegression()
model.fit(X, y)

regression_line = model.predict(X)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(counter, Advanced["PPFGA"])
plt.title("PPFGA Over Time")
plt.xlabel("Game Number")
plt.ylabel("PPFGA")
plt.plot(counter, regression_line)
for i in range(len(Advanced)):
    plt.text(counter[i] + 0.05, Advanced['PPFGA'].iloc[i], 
             Advanced['Opponent'].iloc[i], fontsize=9)
plt.ylim(0.5,1.5)
plt.grid(True)
plt.show()
```
