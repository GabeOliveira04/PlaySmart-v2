```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("DCD.csv", header = None)
data
```

```python
data.columns = data.iloc[1]
data.rename(columns = {"Games": "Period"}, inplace = True)
data = data.drop([0,1,3])
data.drop(["GP", 0.0], axis = 1, inplace = True)
data
```

```python
game = data[:1].reset_index(drop = True)
quarters = data[2:].reset_index(drop = True)
game.iloc[0,0] = "DCD"
game
```

```python
game.to_csv("DCD_game_data.csv", index = False)
quarters.to_csv("DCD_quarters_data.csv", index = False)
```
